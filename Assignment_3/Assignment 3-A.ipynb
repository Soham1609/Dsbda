```python
import numpy as np
import pandas as pd
```

```python
ds=pd.read_csv("Salary Dataset.csv")
ds.info()
```

```python
ds
```

```python
s=set(ds["Location"])
s
```

```python
# Conversion of data from object to string and then remving non-numeric parts of string to empty string
# and finally the string into int
dat1=ds.convert_dtypes()
dat1['Salary'] = dat1['Salary'].str.replace(r'\D', '')
dat1['Salary'] = dat1['Salary'].astype("int64")
dat1.info()
dat1
```

```python
mean_Location= dat1.groupby("Location")["Salary"].mean()
print("Mean according to Locations",mean_Location)
```

```python
median_Location=dat1.groupby("Location")["Salary"].median()
print("Median according to Locations",median_Location)
```

```python
stat_Location=dat1.groupby("Location")["Salary"].describe()
stat_Location
```
